```python
# import graphviz
from graphviz import Digraph
from typing import List, Tuple


def isReflexive(array: List[Tuple]) -> bool:
    inputSet = set()
    for x, y in array:
        inputSet.add(x)
        inputSet.add(y)

    for node in inputSet:
        if (node, node) not in array:
            return False
    return True


def isSymmetric(array: List[Tuple]) -> bool:
    for x, y in array:
        if (y, x) not in array:
            return False
    return True


def isTransitive(array: List[Tuple]) -> bool:
    for x_1, y_1 in array:
        for x_2, y_2 in array:
            if y_1 == x_2 and (x_1, y_2) not in array:
                return False
    return True


def createGraph(array: List[Tuple]) -> Digraph:
    graph = Digraph()
    for x, y in array:
        graph.edge(str(x), str(y))
    return graph


def main(array: List[Tuple]) -> Digraph:
    print("Reflexive: ", isReflexive(array))
    print("Symmetric: ", isSymmetric(array))
    print("Transitive: ", isTransitive(array))

    return createGraph(array)
```

```python
# Test 1
example_array = [
    (0, 0),
    (0, 1),
    (0, 3),
    (1, 0),
    (1, 1),
    (2, 2),
    (3, 0),
    (3, 3),
]
main(example_array)
```
